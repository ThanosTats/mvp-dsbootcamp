```python
#import dataset ratings
ratings= pd.read_csv('/project_data/data_asset/ratings.csv')
print(ratings.head())
print(ratings.shape)
```

```python
ratings.isnull().sum()
```

```python
ratings_gb = ratings.groupby('userId')['movieId'].count().reset_index()

import matplotlib.pyplot as plt
plt.figure(figsize=(25,10))
plt.hist(x=ratings_gb['movieId'])
plt.show()
```
