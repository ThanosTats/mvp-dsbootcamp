# Dataset 2 : movies.csv

```python
df_data_2.describe(include = 'all')
```

```python
#parse genres to seperate columns
df_data_2_T1=df_data_2["genres"].str.split("|",expand=True)
df_data_2_T1.head(5)
```

```python
df_data_2_T1[0].value_counts()
```

```python
# grouping data based on genres
df_data_2_T2 = pd.get_dummies(df_data_2_T1,prefix='', prefix_sep='')
df_data_2_T2 = df_data_2_T2.groupby(level=0, axis=1).sum()
```

```python
df_data_2_T2.head(5)
```

```python
# df_data_2_T2.groupby(level=0, axis=1).sum().head(5)                                                
```

```python
# Rename column Name
df_data_2_T2.rename({'(no genres listed)':'NoGenresListed'}, axis=1, inplace = True)
df_data_2_T2.head(2)
```

```python
# Get all names 
for col_name in df_data_2_T2.columns: 
    print(col_name)
```

```python
# Concat two data frames 
df_data_2_T3 = pd.concat([df_data_2,df_data_2_T2], axis=1,sort = False)
```

```python
del df_data_2_T3["genres"]
# Delete Column
df_data_2_T3.head()
```
